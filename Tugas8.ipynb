```python
def cek_kelulusan(nilai):
    if nilai < 60:
        return "Gagal"
    elif nilai >= 61 and nilai <= 70:
        return "Baik"
    elif nilai >= 71 and nilai <= 80:
        return "Sangat Baik"
    elif nilai >= 81 and nilai <= 100:
        return "Istimewa"
    else:
        return "Nilai tidak valid"

# Contoh penggunaan
print(cek_kelulusan(nilai=65))
```

```python
def ganjil(n):
    for i in range(1, n+1, 2):
        print(i)

ganjil(100)
```
